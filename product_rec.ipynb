```python
import numpy as np 
import pandas as pd 
```

```python
df['Ingredients']=df['Ingredients'].apply(lambda x:x.lower())
df.head()
```

```python
df=df.copy()
```

```python
df=df.drop(columns=['Name','Price','Rank','Combination','Dry','Normal','Oily','Sensitive'],axis=1)
df.head()
```

```python
from nltk.stem.porter import PorterStemmer 
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
        
    return  " ".join(y)
```

```python
df['Ingredients']=df['Ingredients'].apply(stem)
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(df['Brand']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
# sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6] #sorting based on number
```

```python
def recommend(product):
    product_index=df[df['Brand'] == product].index[0]
    distances=similarity[product_index]
    product_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in product_list:
        # print(df.iloc[0].Brand)
        # print(df.iloc[0].Ingredients)
        # print(i[0])
        print(i)
```

```python
recommend("FRESH")
```
